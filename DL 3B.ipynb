```python
x_train = x_train/255
x_test = x_test/255
```

```python
 x_train = x_train.reshape(1887, 28,28)
 x_test = x_test.reshape(1887, 28, 28, 1)
```

```python
print("Train Shape :",x_train.shape)
print("Test Shape :",x_test.shape)
print("y_train shape :",y_train.shape)
print("y_test shape :",y_test.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPooling2D, Flatten
```

```python
model=Sequential()
model.add(Conv2D(64, (3,3), activation='relu', input_shape=(28,28,1)))
model.add(MaxPooling2D((2,2)))
model.add(Conv2D(64, (3,3), activation='relu'))
model.add(MaxPooling2D((2,2)))
model.add(Flatten())
model.add(Dense(128,activation='relu'))
model.add(Dense(10,activation='softmax'))
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
model.summary()
```

```python
model.fit(x_train, y_train, epochs=3, verbose=1,validation_data=(x_test,y_test))
```

```python
predictions = model.predict(x_test)
```

```python
 import numpy as np
 index=10
 print(predictions[index])
 final_value=np.argmax(predictions[index])
 print("Actual label :",y_test[index])
 print("Predicted label :",final_value)
 print("Class :",class_names[final_value])
```

```python
 plt.imshow(x_test[10])
```

```python
loss, accuracy = model.evaluate(x_test, y_test)
print("Loss :",loss)
print("Accuracy (Test Data) :",accuracy*100)
```
